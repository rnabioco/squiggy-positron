```python
  from squiggy import FastaFile, BamFile

  # Search for DRACH motif (m6A consensus)
  with FastaFile('../squiggy/data/yeast_trna.fa') as fasta:
      matches = list(fasta.search_motif('DRACH'))
      for match in matches:
          print(f"{match.chrom}:{match.position} {match.sequence} ({match.strand})")

  # Find reads overlapping motif positions
  with BamFile('../squiggy/data/yeast_trna_mappings.bam') as bam, FastaFile('../squiggy/data/yeast_trna.fa') as fasta:
      overlaps = bam.get_reads_overlapping_motif(fasta, 'DRACH')
      for position, reads in overlaps.items():
          print(f"{position}: {len(reads)} reads")
```
